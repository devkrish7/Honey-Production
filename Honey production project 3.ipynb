# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
sns.displot(df["priceperlb"]);
plt.title("Average price per lb")
plt.xlabel('Price per lb')
plt.ylabel('Count')
```

Our majority of data is before the mean (1.409569), our median (1.360000) is before the mean, and it has positive skewness. Our majority of data is between the range (0.7 to 2)

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
sns.scatterplot(x=df["numcol"],y=df["prodvalue"]);
plt.title("relationship between the variables 'numcol' and 'prodval'")
```

Because there is linear growth between them, there is a positive link between the number of colonies and the production value.The majority of data falls between 0 and 100,000.

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
sns.boxplot(x=df["year"],y=df["prodvalue"]);
plt.title("relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot")
```

The 2003,2008,2009,2010,2012 has the more outliers and the 2010 has the highest outliers. Almost majority of the years have the same range of mean.

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
sns.pairplot(df,vars=["numcol","yieldpercol","totalprod","prodvalue","year"]);
```

Some of the pairs have no association, such as (yieldpercol,numcol), (prodvalue, yieldpercol), and (yieldpercol,totalprod). These pairs have a positive correlation because as the x axis increases, so does the y axis. Thus the two variables (totalprod,numcol), (prodvalue,numcol), and (totalprod,prodvalue) are all flowing in the same direction.


# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol, 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
sns.heatmap(df[['numcol','yieldpercol', "totalprod", "stocks", "priceperlb", 'prodvalue']].corr(),annot=True)
```

The correlation value is high for the totalprod and numcol(0.95). The least positive correlation is between the totalprod and the yieldpercol(0.4). There is also some negatively correlated pairs they are mostly pair with the priceperlb such as numcol, yieldpercol,totalprod ,prodvalue and stocks .
